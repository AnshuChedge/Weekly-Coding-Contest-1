```python
"""QUE 1: Array Reverse
Reverse an array arr[]. Reversing an array means rearranging the elements such that the first element 
becomes the last, the second element becomes second last and so on."""

arr = list(map(int, input("Enter elements: ").split()))
print("Array is:",arr)
arr.reverse()
print("Reversed array is:",arr)
```

```python
"""QUE 2: Second Largest Element in an Array
Given an array of positive integers arr[] of size n, the task is to find second largest distinct element
in the array.
"""
arr = list(map(int, input("Enter elements: ").split()))
print("Array is:",arr)
arr.sort()
arr.reverse()
print("second largest element of array is:",arr[1])
```

```python
"""QUE 3: Given an integer array nums, return the sum of all elements in the array"""
arr = list(map(int, input("Enter elements: ").split()))
print("Array is:",arr)
sum_arr=sum(arr)
print("Sum of array is:",sum_arr)
```

```python
"""QUE 5 : Search an element in an array
You are given an array A of size N and an element X. Your task is to find whether the array A contains the element X or not.
"""
A=list(map(int, input("Enter elements: ").split()))
X=int(input("Enter the element:"))
print("Array is:",A)
print("element to be searched is:",X)
for i in A:
    if i == X:
        print("Element found")
        break
else:
    print("Element not found")
```

```python
"""QUE 6: PRINT A 2D ARRAY
Create a 2d array and print it. 
"""
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

A = []

for i in range(rows):
    row = []
    for j in range(cols):
        row.append(int(input(f"Enter element [{i}][{j}]: ")))
    A.append(row)

for row in A:
    print(row)
```
