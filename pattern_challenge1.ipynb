```python
##1.n Ã— n Square Pattern (Space-Separated Stars) 
n=int(input("Enter the number:"))
for i in range(n):
    print(" * " * n)
```

```python
##2.Left-Aligned Increasing Triangle (Stars, Space-Separated)
n=int(input("Enter the number:"))
for i in range(1,n+1):
    print(" * " * i)
```

```python
##3. Decreasing Star Triangle (Space-Separated)
n=int(input("Enter the number:"))
for i in range(n,0,-1):
    print(" * " * i)
```

```python
##4. Decreasing Star Triangle (Full width)
n=int(input("Enter the number:"))
for i in range(0, n):
    stars = 2*(n-i)-1
    spaces = i
    print(" "*spaces + "*"*stars)

```

```python
##5.  Number Triangle (1 to i in Each Row) 
n=int(input("Enter the number:"))
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print("")
```

```python
##6. Repeating Number Triangle
n=int(input("Enter the number:"))
for i in range(1,n+1):
    for j in range(1,i+1):
        print(i,end=" ")
    print("")
```

```python
##7.Reverse Number Triangle (1 to decreasing upper limit)
n=int(input("Enter the number:"))
for i in range(n,0,-1):
    for j in range(1,i+1):
        print(j,end=" ")
    print("")
```

```python
##8. Symmetric Star Diamond (Vertical Expansion + Contraction)
n=int(input("Enter the number:"))
for i in range(1,n+1):
    print(" * " * i)
for i in range(n-1,0,-1):
    print(" * " * i)
```

```python
##9. Hollow Square (Stars)
n=int(input("Enter the number:"))
for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
##10. Alphabet Triangle (Repeated Letters) 
n = int(input("Enter the number: "))
for i in range(1, n+1):
    ch = chr(64 + i)     
    for j in range(1, i+1):
        print(ch, end=" ")
    print()
```

```python
##11. Reverse Alphabet Triangle
n = int(input("Enter the number: "))
for i in range(n, 0, -1):    
    for j in range(i):
        print(chr(65 + j), end="")
    print(" ")
```

```python
##12. Alphabet Palindrome Triangle 
n = int(input("Enter number of rows: "))
for i in range(1, n + 1):
    print(" " * (n - i), end="")
    for j in range(i):
        print(chr(65 + j), end="")
    for j in range(i - 2, -1, -1):
        print(chr(65 + j), end="")
    print()
```

```python
#13. Concentric Number Square 
n = int(input("Enter a number: "))
size = 2 * n - 1   
for i in range(size):
    for j in range(size):
        top = i
        left = j
        right = size - 1 - j
        bottom = size - 1 - i
        layer = n - min(top, left, right, bottom)
        print(layer, end=" ")
    print()
```

```python
##15. X-Shaped Star Pattern (Growing + Shrinking)
n=int(input("Enter a Number:"))
for i in range(1, n+1):
    print("*"*i + " "*(2*(n-i)) + "*"*i)

for i in range(n-1, 0, -1):
    print("*"*i + " "*(2*(n-i)) + "*"*i)
```
